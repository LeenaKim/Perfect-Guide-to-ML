# 빅분기 3회 기출 - 작업형2유형 복기

```python
import pandas as pd

df = pd.read_csv('archive/TravelInsurancePrediction.csv')
```

```python
df.info()
```

```python
x_train = df.iloc[:, 1:8]
y_train = df.iloc[:, 9]
x_train.info()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_train, y_train, test_size=0.2, random_state=101)
```

```python
x_train_dum = pd.get_dummies(x_train)
x_test_dum = pd.get_dummies(x_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(x_train_dum)
x_train_scl = sc.transform(x_train_dum)
x_test_scl = sc.transform(x_test_dum)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_clf = RandomForestClassifier(max_depth=5, random_state=111)
rf_clf.fit(x_train_scl, y_train)
pred = rf_clf.predict_proba(x_test_scl)
```

```python
pred
```

```python
pred_1 = pd.Series(pred[:, 1])
```

```python
pred_1
```

```python
result = pd.DataFrame(pred_1, columns=['y_pred'])
result.to_csv('result2.csv', index_label='index')
```

```python
from sklearn.metrics import accuracy_score

pred2 = rf_clf.predict(x_test_scl)
accuracy = accuracy_score(y_test, pred2)
print('랜덤포레스트 정확도 : {0:.4f}'.format(accuracy))
```

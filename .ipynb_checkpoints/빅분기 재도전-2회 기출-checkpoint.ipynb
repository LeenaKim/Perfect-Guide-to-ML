# 빅분기 2회차 기출

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestClassifier
```

```python
x_train = pd.read_csv('archive/X_train.csv')
y_train = pd.read_csv('archive/y_train.csv')
x_test = pd.read_csv('archive/X_test.csv')
```

```python
x_train.info()
```

```python
y_train.info()
```

```python
# null값은 없으니 train 데이터의 id 값을 빼고 원핫인코딩 수행 
x_train = x_train.drop(['ID'], axis=1)
x_dummies = pd.get_dummies(x_train)

x_test_dropped = x_test.drop(['ID'], axis=1)
test_dummies = pd.get_dummies(x_test_dropped)
```

```python
x_dummies.head(10)
```

```python
y_train.head(10)
```

```python
# 스케일러로 전처리 
scaler = MinMaxScaler()
scaler.fit(x_dummies)
train_scaled = scaler.transform(x_dummies)
test_scaled = scaler.transform(test_dummies) # 학습데이터로 스케일한 객체로 테스트 데이터도 스케일링 
```

```python
# 랜덤포레스트 분류모형 
y_train = y_train['Reached.on.Time_Y.N']
rf_clf = RandomForestClassifier(max_depth=5, random_state=0)
rf_clf.fit(train_scaled, y_train)
```

```python
# 제 시간에 올 확률 예측 
from sklearn.metrics import classification_report

y_pred = rf_clf.predict_proba(test_scaled)
```

```python
y_pred
```

```python
# y_pred의 첫번째 칼럼은 0일 확률, 두번째 칼럼은 1일 확률(제시간에 도착)
# 엑셀화 
y_test = pd.Series(y_pred[:, 1])
result = pd.concat([x_test.ID, y_test], axis=1)
```

```python
result
```

```python
# 칼럼명 변경 
result.columns = ['ID', 'result']
#result.to_csv("archive/result.csv", index=False)
```
